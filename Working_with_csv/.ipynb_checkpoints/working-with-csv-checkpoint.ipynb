# 1. Importing pandas

```python
import pandas as pd
```

# 2. Opening a local csv file
pd.read_csv is a function in the Pandas library used to read CSV (Comma-Separated Values) files and load them into a DataFrame, which is a 2D data structure like a table.

```python
df = pd.read_csv('aug_train.csv')
df
```

# 3. Opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# 4. Sep Parameter
The sep parameter in pd.read_csv() specifies the delimiter — the character that separates columns in your CSV file.

```python
pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])
```

# 5. Index_col parameter
The index_col parameter in pd.read_csv() tells Pandas which column(s) to use as the row labels (index) of the resulting DataFrame.

```python
pd.read_csv('aug_train.csv',index_col='enrollee_id')
```

# 6. Header parameter
The header parameter in pd.read_csv() tells Pandas which row to use as the column names (i.e., the header row) when reading a CSV file.

```python
pd.read_csv('test.csv')
```

```python
pd.read_csv('test.csv',header= 1)
```

# 7. use_cols parameter
The usecols parameter in pd.read_csv() is used to load only specific columns from a CSV file — which helps reduce memory usage and speeds up processing when you don’t need all the columns.

```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])
```

# 8. Skiprows/nrows Parameter
**skiprows-** The skiprows parameter in pd.read_csv() is used to skip rows at the start of the file when reading a CSV.


```python
pd.read_csv('aug_train.csv',skiprows=[0,2])
```

**nrows-** used to limit the number of rows read from the CSV file.

```python
pd.read_csv('aug_train.csv',nrows=100)
```

# 10. Encoding parameter
The **encoding** parameter in pd.read_csv() specifies the character encoding used to read the CSV file correctly.

```python
pd.read_csv('zomato.csv',encoding='latin-1')
```

# 11. on_bad_lines
The **on_bad_lines** parameter in **pd.read_csv()** is used to handle malformed or problematic rows in a CSV file — such as rows with too many or too few columns.

```python
pd.read_csv('zomato.csv',encoding="latin-1",on_bad_lines='skip')
```

# 12. dtype parameter
The **dtype** parameter in **pd.read_csv()** lets you explicitly set the data type for one or more columns while reading a CSV file.

```python
pd.read_csv('aug_train.csv').info()
```

```python
# converting target column's datatype from float to interger
pd.read_csv('aug_train.csv',dtype={'target':int}).info()
```

# 13. Handling Dates
Handling dates in **pd.read_csv()** is done using the **parse_dates** and optionally **dayfirst** or **date_parser** parameters.

```python
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

# 14. Convertors

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

# 15. na_values parameter
The **na_values** parameter in pd.read_csv() is used to specify custom missing value indicators — i.e., values that should be interpreted as NaN (missing) when reading the CSV file.

```python
pd.read_csv('aug_train.csv',na_values=['Male']) # in this parameter we can also pass more than one values
```

# 16. Loading a huge dataset in chunks

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunk in dfs:
    print(chunk.shape)
```
